8. Ada sebuah teater yang memiliki 10 baris. Jika baris pertama berisi 9 kursi dan baris kedua berisi 13 kursi, berapakah kursi pada baris ke-9? Hitunglah menggunakan Python dan menggunakan rumus barisan aritmatika. Jelaskan setiap step yang kamu lakukan dengan menggunakan comment.

  image.png

```python
# simpan parameter ke dalam variabel
un = 9
uDua = 13
n=10

# hitung menggunakan rumus, simpan dalam variabel barisKeN
barisKeN = un + (n-1) * (uDua - un)

# print barisKeN
print("kursi pada baris ke-{0}, adalah: {1}".format(n, barisKeN))
```

9. Sebuah dadu dilempar satu kali. Tentukan peluang ketika:

  a. Kejadian A munculnya mata dadu dengan angka prima

  b. Kejadian munculnya mata dadu dengan jumlah kurang dari 6

  Hitunglah menggunakan Python dan menggunakan rumus peluang. Berilah penjelasan untuk setiap yang kamu lakukan dalam comment.
  image.png

```python
# buat function prima
def prima(array):
    kejadianPrima = 0
    for i in array:
        now = 1
        factor = 0
        while now <= i:
            if i%now == 0:
                factor += 1
            now += 1
        if factor == 2:
            kejadianPrima += 1
    return kejadianPrima

# function kurang dari 6
def kurang(array):
    kejadian = 0
    for i in array:
        if i<6:
            kejadian+=1
    return kejadian

# masukkan kejadian dalam variabel
dadu = [1, 2, 3, 4, 5, 6]
kejadianPrima = prima(dadu)
kejadianB = kurang(dadu)
peluangA = kejadianPrima/len(dadu)
peluangB = kejadianB/len(dadu)
print("peluang kejadian munculnya mata dadu dengan angka prima: {}".format(peluangA))
print("peluang kejadian munculnya mata dadu dengan jumlah kurang dari 6: {}".format(peluangB))
```
